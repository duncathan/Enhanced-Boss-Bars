```python
from PIL import Image  # if not installed, do : pip install pillow

import glob  # file system iterator

offset = -5
folderToCalculate = "resources/gfx/ui/*/final/"

files = glob.glob(folderToCalculate + '/*.png', recursive=False)
for filename in files:
    print(filename)
    im = Image.open(filename)
    pix = im.load()

    startPos = [im.size[0] / 2, im.size[1] / 2]
    pixVal = pix[startPos[0], startPos[1]]
    stepsTaken = offset
    while pixVal[3] == 255 and startPos[0] + 1 < im.size[0]:
        startPos[0] += 1
        pixVal = pix[startPos[0], startPos[1]]
        stepsTaken += 1
    print("Vector(-" + str(stepsTaken) + ", 0)")
```
